```python
# Import the necessary libraries

import qiskit as q
import matplotlib.pyplot as plt 

# Phi + Bell State:

# Create a simple quantum circuit using 2 qubits and 2 classical bits

qr = q.QuantumRegister(2) # create register to store bits
cr = q.ClassicalRegister(2)
circuit = q.QuantumCircuit(qr, cr) # creates the circuit 

# Apply the Hadamard gate on the 0th qubit

circuit.h(qr[0]) # 0th index on the quantum register 

# Apply the CNOT gate with the output of Hadamard gate as control bit

circuit.cx(qr[0], qr[1]) # circuit.cx(control_bit, target_bit)

# Measure

circuit.measure(qr, cr) # measure quantum bit into classical bit
circuit.draw(output='mpl')

```

```python
# The probability amplitude for the states |00⟩ and |11⟩ is 1/2 and there is 0% chance of measuring the states |01⟩ and |10⟩.
# Simulate locally

simulator = q.Aer.get_backend('qasm_simulator') 
results = q.execute(circuit, backend=simulator).result()
q.visualization.plot_histogram(results.get_counts(circuit))
```
